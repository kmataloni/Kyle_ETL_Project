#### World Wholesale Meat Prices


```python
World_Wholesale_Meat_Prices = "../ETL Project/WholesalePrices.csv"

PricesDF = pd.read_csv(World_Wholesale_Meat_Prices)
```

```python
PricesDF
```

```python
PricesDF.info()
```

```python
#Rename first 5 columns
PricesDF.rename(columns={
    "Unnamed: 0": "Date",
    "Beef, Central U.S.": "Select Boxed Beef",
    "Unnamed: 2": "Choice Boxed Beef",
    "Unnamed: 3": "Boneless Beef",
    "Unnamed: 4": "Imported Boneless Beef"
    }, inplace=True)
PricesDF.head()
```

```python
#Remove headers from DF - only keep column names
NewPricesDF = PricesDF.iloc[3:]
NewPricesDF
```

```python
#Take only the first 5 columns, with all rows
NewPricesDF2 = NewPricesDF.iloc[:,:5]
NewPricesDF2
```

```python
NewPricesDF2
```

```python
NewPricesDF3 = NewPricesDF2.copy()
```

```python
#Re-cast as a date
NewPricesDF3["Date"] = pd.to_datetime(NewPricesDF3["Date"], format = "%b-%y")
NewPricesDF3
```

```python
#Re-cast remaining columns as floats
NewPricesDF3 = NewPricesDF3.astype({
    "Select Boxed Beef": "float", 
    "Choice Boxed Beef": "float",
    "Boneless Beef": "float",
    "Imported Boneless Beef": "float"
})
NewPricesDF3.info()
```

```python
PricesByYear = NewPricesDF3.groupby([NewPricesDF3["Date"].dt.year]).mean()
PricesByYear
```

```python
PricesByYear["Avg"] = PricesByYear[["Select Boxed Beef", "Choice Boxed Beef", "Boneless Beef", "Imported Boneless Beef"]].mean(axis=1)
PricesByYear
```

```python
NewDf = pd.DataFrame(PricesByYear["Avg"]).round(2)
NewDf
```

```python
Final_Consumption_DF2 = Final_Consumption_DF.reset_index(drop=True).round(2)
Final_Consumption_DF2
```

```python
Final_Consumption_DF2.set_index("Date")
```

```python
Final_Consumption_DF3 = Final_Consumption_DF2.iloc[:,3:5].set_index("Date")
Final_Consumption_DF3
```

```python
mergedDF = NewDf.merge(Final_Consumption_DF3, left_index=True, right_index=True)
mergedDF
```

```python
mergedDF.rename(columns={
    "Avg": "Average Price of Beef",
    "Tons": "Beef Intake (kg/cap)"
    },inplace=True)
mergedDF
```

```python
mergedDF.to_csv("../ETL Project/BeefData.csv")
```
